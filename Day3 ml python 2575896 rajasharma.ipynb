Separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
y = data_frame['label']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Build and train the initial logistic regression model
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Make predictions
y_pred = log_reg.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1 Score:', f1)
```

```python
# Feature engineering techniques such as creating new features, scaling, handling missing values, and encoding categorical variables can be applied here.
# Example: Scaling numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

b. Explain the relevance of each feature engineering technique:
Creating New Features: New features can capture complex patterns in the data, improving the model's predictive power.
Scaling or Normalizing Features: Scaling ensures that all features have the same influence on the model, preventing some features from dominating due to their larger scale.
Handling Missing Values: Missing values can lead to biased predictions. Techniques like imputation or removing missing data points can be used.
Encoding Categorical Variables: Logistic regression models require numerical inputs. Encoding converts categorical variables into numerical format, making them usable by the model.

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

# Apply SMOTE (oversampling) to the minority class
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_scaled, y_train)

# Apply random undersampling to the majority class
rus = RandomUnderSampler(random_state=42)
X_train_resampled, y_train_resampled = rus.fit_resample(X_train_resampled, y_train_resampled)
```

```python
# Build and train logistic regression model with feature-engineered and balanced data
log_reg_balanced = LogisticRegression()
log_reg_balanced.fit(X_train_resampled, y_train_resampled)

# Make predictions
y_pred_balanced = log_reg_balanced.predict(X_test_scaled)

# Evaluate the model
accuracy_balanced = accuracy_score(y_test, y_pred_balanced)
precision_balanced = precision_score(y_test, y_pred_balanced)
recall_balanced = recall_score(y_test, y_pred_balanced)
f1_balanced = f1_score(y_test, y_pred_balanced)

print('Accuracy (Balanced):', accuracy_balanced)
print('Precision (Balanced):', precision_balanced)
print('Recall (Balanced):', recall_balanced)
print('F1 Score (Balanced):', f1_balanced)
```

```python
# Get coefficients and corresponding feature names
coefficients = log_reg_balanced.coef_[0]
feature_names = X.columns

# Create a DataFrame to store coefficients and feature names
coef_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})

# Sort the DataFrame by coefficient values to identify influential features
coef_df_sorted = coef_df.sort_values(by='Coefficient', ascending=False)

print(coef_df_sorted)
```
